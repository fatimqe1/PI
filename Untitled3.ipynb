```
# -*- coding: utf-8 -*-
"""Proyecto_final_POOE.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1drGuPGzkuv7YY7JrRXR7w4l0PkT1k8CM
"""

#Proyecto:Simulacion de mareas para ver la potencia que produce una turbina

#Librerias
import  numpy             as np 
import  matplotlib.pyplot as plt
import  scipy             as sp
import  matplotlib.cm     as cm
import  pandas            as pd
import  math

#Se crea una clase para los atributos de la marea
class Marea:
    #MÃ©todos
    def __init__(self,pl,D,A1,A2,A3,A4):
        self.playa     =pl
        self.dia       =D
        self.al1       =A1
        self.al2       =A2
        self.al3       =A3
        self.al4       =A4
        
class turbina:
    def __init__(self,pla,d,U):
        self.playa    =pla
        self.dias     =d
        self.densidad =1025
        self.area     =50.26 #"""Diametro de la turbina=15m"""
        self.velocidad=U
        
#Lectura de los archivos
pv    = pd.read_csv('marea - puertov.csv')
qp   = pd.read_csv('marea - quepos.csv')
ml    = pd.read_csv('marea - MLouis.csv')
en    = pd.read_csv('marea - ensenada.csv')

"""*************AsignaciÃ³n de valores a mareas**************"""
"""Marea primaver"""
Mr1    =Marea('Pueto vallarta',pv['dia'], pv['amp1'], pv['amp2'],pv['amp3'],pv['amp4'])
"""Marea verano"""
Mr2    =Marea('Quepos',qp['dia'],qp['amp1'],qp['amp2'],qp['amp3'],qp['amp4'])
"""Marea otoÃ±o"""
Mr3    =Marea('MLouis',ml['dia'],ml['amp1'],ml['amp2'],ml['amp3'],ml['amp4'])
"""Marea invierno"""
Mr4    =Marea('Ensenada',en['dia'],en['amp1'], en['amp2'],en['amp3'],en['amp4'])

#print(Mr1.al1)

   
"""**********Velocidad de corriente*****"""""
#Se define la funcion para graficar la corriente
def velocidad(ply,dia,alt1,alt2,alt3,alt4):
   #Con la formula 
   #$U_{(t)}=\sum A_{i}cos(\boldsymbol{\omega_{i}t+\boldsymbol{\varphi_{i}}})$
   #Creacion de la funcion para la sumatoria, 
   #este parametro es igual para todos los mares
    cosen = math.cos(10*math.pi+(math.pi/8))
    U =((cosen*alt1)+(cosen*alt2)+(cosen*alt3)+(cosen*alt4))/(24*1200)
    print(U)
    plt.title('Velocidad de la marea')
    plt.ylabel('Velocidad m/s')
    plt.xlabel('Tiempo (dÃ­as)')
    plt.subplot(1,2,1)
    plt.plot(dia, U, '-o', label=ply)
    plt.show()   
    return U

#Se imprimen las graficas utilizando la funcion
print('Puerto Vallarta')
U1=velocidad('Puerto Vallarta',Mr1.dia, Mr1.al1, Mr1.al2, Mr1.al3,Mr1.al4)
print('Quepos')
U2=velocidad('Quepos',Mr2.dia, Mr2.al1, Mr2.al2, Mr2.al3,Mr2.al4)
print('Mlouis')
U3=velocidad('MLouis',Mr3.dia, Mr3.al1, Mr3.al2, Mr3.al3,Mr3.al4)
print('Ensenada')
U4=velocidad('Ensenada',Mr4.dia, Mr4.al1, Mr4.al2, Mr4.al3,Mr4.al4)


"""**********CreaciÃ³n de los objetos de turbina**********"""
Tr1  =turbina('Puerto Vallarta',Mr1.dia,U1)
Tr2  =turbina('Quepos',Mr2.dia,U2)
Tr3  =turbina('MLouis',Mr3.dia,U3)
Tr4  =turbina('Ensenada',Mr4.dia,U4)


"""*******CÃ¡lculo de la potencia******"""
def potencia(ply,dia,de,a,v):
    #Formula: Potencia=1/2*dAV^3 donde d=densidad,A=area de la turbina y V=velocidad
    P=0.5*de*a*v*v*v
    print(P)
    plt.subplot(133)
    plt.title('Potencia de la marea')
    plt.xlabel('Tiempo(Dias)')
    plt.ylabel('Potencia (W)')
    plt.subplot(1,2,2)
    plt.plot(dia,P,'-o',label=ply)
    plt.legend()
    plt.show()   
    return P



print('Potencia en PVallarta')
P1=potencia(Tr1.playa,Tr1.dias,Tr1.densidad,Tr1.area,Tr1.velocidad)
print('Potencia en Quepos')
P2=potencia(Tr2.playa,Tr2.dias,Tr2.densidad,Tr2.area,Tr2.velocidad)
print('Potencia en MLouis')
P3=potencia(Tr3.playa,Tr3.dias,Tr3.densidad,Tr3.area,Tr3.velocidad)
print('Potencia en Ensenada')
P4=potencia(Tr4.playa,Tr4.dias,Tr4.densidad,Tr4.area,Tr4.velocidad)
```
